            int[] numberArray = new int[n];

            Console.WriteLine($"Input {n} elements in the array :");
            for (int i = 0; i < numberArray.Length; i++)
            {
                numberArray[i] = int.Parse(Console.ReadLine());
            }

            List<int> oddNumbers = new List<int>();
            List<int> evenNumbers = new List<int>();

            //seperate even/odd numbers 
            foreach (int num in numberArray)
            {
                if (num % 2 == 0)
                {
                    evenNumbers.Add(num);
                }
                else
                {
                    oddNumbers.Add(num);
                }
            }

            //store even/odd numbers in seperate arrays
            int[] evenNumbersArray = evenNumbers.ToArray();
            int[] oddNumbersArray = oddNumbers.ToArray();

            Console.WriteLine("The Even elements are:");
            foreach (int evenNum in evenNumbersArray)
            {
                Console.Write(evenNum + " ");
            }

            Console.WriteLine();
            Console.WriteLine("The Odd elements are :");
            foreach (int oddNum in oddNumbersArray)
            {
                Console.Write(oddNum + " ");
            }
        }
    }

# Answer for Question 6 Python

```python
consonants = ('b','c','d', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'y', 'z')
vowels = ('a', 'e', 'i', 'o', 'u')
way = "way"
ay = "ay"

userInput = input("Enter a word to translate into Pig Latin word: ")

# gettinjg input and make sure it is a string, then remove white spaces at the begining/end of the word if any
stripedWord = str(userInput).strip();

# convert it to lowercas
stripedWord = stripedWord.lower()

# getting first letter
firstletter = stripedWord[0]

pigLatinWord = ""

if firstletter in consonants:
    #remove first letter
    stripedWord = stripedWord.replace(stripedWord[0], '')
    
    #converting to piglatin
    pigLatinWord = stripedWord + firstletter + ay
elif firstletter in vowels:
    #converting to piglatin
    pigLatinWord = stripedWord + way
    
print(pigLatinWord)
```

# Answer for Question 10 Python

```python
def frequencies(text):
    charFrequencies = {}
    letters = "abcdefghijklmnopqrstuvwxyz"
    
    for char in text:
        if (char.isalpha()) == True:
            char = char.lower()
            if char in charFrequencies:
                charFrequencies[char] += 1
            else:
                charFrequencies[char] = 1
    
    listOfCharFrequesncies = []
    for letter in letters:
        if letter in charFrequencies:
            listOfCharFrequesncies.append(charFrequencies[letter])
        else:
            listOfCharFrequesncies.append(0)
    print (listOfCharFrequesncies)
   

text1 = "The quick red fox got bored and went home."
text2 = "apple"
print("freuencies of 'The quick red fox got bored and went home.'")
frequencies(text1)

print("freuencies of 'apple'")
frequencies(text2)
```
